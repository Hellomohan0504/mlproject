```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, BatchNormalization, LeakyReLU
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
data = pd.read_csv('hvac_dataset.csv')
data = data.dropna()
data[["Date", "Time"]] = data["Date"].str.split(" ", expand=True)
data['target'] = data["outlet_temp"].shift(-1)
data = data.drop(data.tail(1).index)
```

```python
n = len(data)
train_df = data[0:int(n*0.7)]
val_df = data[int(n*0.7):int(n*0.9)]
test_df = data[int(n*0.9):]

x_train = train_df.drop("outlet_temp", axis=1)
x_val = val_df.drop("outlet_temp", axis=1)
x_test = test_df.drop("outlet_temp", axis=1)

y_train = train_df["outlet_temp"]
y_val = val_df["outlet_temp"]
y_test = test_df["outlet_temp"]
```

```python
x_train_numeric = x_train.drop(columns=['Date', 'Time'])
x_val_numeric = x_val.drop(columns=['Date', 'Time'])
x_test_numeric = x_test.drop(columns=['Date', 'Time'])

# Standardize features
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train_numeric)
x_val_scaled = scaler.transform(x_val_numeric)
x_test_scaled = scaler.transform(x_test_numeric)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train_numeric.columns, index=x_train.index)
x_val_scaled = pd.DataFrame(x_val_scaled, columns=x_val_numeric.columns, index=x_val.index)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test_numeric.columns, index=x_test.index)

# Prepare data for LSTM (sequence length)
Seq_len = 5  # Sequence length
batch_size = 32  # Batch size

def create_sequences(data, target, seq_len=10):
    sequences = []
    labels = []
    for i in range(len(data) - seq_len):
        sequences.append(data[i:i+seq_len].values)
        labels.append(target[i+seq_len])
    return np.array(sequences), np.array(labels)
```

```python
# Ensure that the target series (y_train, y_val, y_test) have the same index as the feature data
y_train = y_train.reset_index(drop=True)
y_val = y_val.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

# Now create the sequences
x_train_seq, y_train_seq = create_sequences(x_train_scaled, y_train, Seq_len)
x_val_seq, y_val_seq = create_sequences(x_val_scaled, y_val, Seq_len)
x_test_seq, y_test_seq = create_sequences(x_test_scaled, y_test, Seq_len)
```

```python
# LSTM model
input_shape = (Seq_len, x_train_seq.shape[2])  # (sequence_length, number_of_features)

model = Sequential([
    Input(shape=input_shape),
    LSTM(64, return_sequences=True),
    LSTM(32),
    Dense(128),
    LeakyReLU(alpha=0.1),
    Dropout(0.5),
    BatchNormalization(),
    Dense(1)  # Linear output layer for regression
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
early_stopping_cb = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Training the model
history = model.fit(x_train_seq, y_train_seq, 
                    epochs=100, 
                    batch_size=batch_size, 
                    validation_data=(x_val_seq, y_val_seq),
                    callbacks=[early_stopping_cb])
```

```python
test_loss = model.evaluate(x_test_seq, y_test_seq)
print(f'Test loss: {test_loss}')

# Predict
y_pred = model.predict(x_test_seq)
```

```python
# Compute additional metrics
r2 = r2_score(y_test_seq, y_pred.flatten())
mae = mean_absolute_error(y_test_seq, y_pred.flatten())
mse = mean_squared_error(y_test_seq, y_pred.flatten())

# Print the metrics
print(f"R² Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
```
